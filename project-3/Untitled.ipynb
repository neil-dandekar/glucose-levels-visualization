```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('full.csv')
df
```

```python
# Plot 1: Relationship between Average Meal Calories and Glucose Spike
plt.figure(figsize=(8, 5))
plt.scatter(df["Average_Meal_Calories"], df["Average_Glucose_Spike"], alpha=0.5)
plt.xlabel("Average Meal Calories")
plt.ylabel("Average Glucose Spike")
plt.title("Glucose Spike vs. Meal Calories")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.boxplot(column="BMI", by="Diabetes_Status", grid=False)
plt.xlabel("Diabetes Status")
plt.ylabel("BMI")
plt.title("BMI Distribution by Diabetes Status")
plt.suptitle("")  # Remove default title
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df[["Carbs_Percentage", "Protein_Percentage", "Fat_Percentage", "Fiber_Percentage"]].mean().plot(kind="bar")
plt.ylabel("Average Proportion")
plt.title("Macronutrient Proportions in Meals")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```
